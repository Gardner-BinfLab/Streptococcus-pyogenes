```python
protein = pd.read_csv('protein.txt.gz', sep='\t', header=None)
protein['Accession'] = protein[1].apply(lambda x: x.split(' ')[0])
protein['Multispecies'] = protein[1].apply(lambda x: x.split(' ')[1])
protein['Description'] = protein[1].apply(lambda x: ' '.join(x.split(' ')[2:]))
protein = protein.rename(columns={0:'GCA', 2:'Protein'})
protein = protein.drop(1, axis=1)
protein.head()
```

```python
eggnog = pd.read_csv('../doc/protein.99.eggnog.txt', sep='\t', header=None)
eggnog.columns = ['Accession', 'eggNOG', 'E-value', 'Bitscore']
eggnog.head()
```

```python
new_df = pd.merge(protein, eggnog, on='Accession')
new_df.head()
```

```python
df1 = pd.merge(new_df, \
         df[['GCA', 'emm Family', 'emm Sub-type', 'tissue/source', 'Clinical', 'Invasive (pyseer)',]],\
        on='GCA')
df1.head()
```

```python


# ax = sns.boxplot(data=df1, x='Clinical', y='Bitscore')
# ax.set_yscale('log')
# plt.xticks(rotation=45, ha='right');


cols = ['emm Family', 'emm Sub-type', 'tissue/source', 'Clinical', 'Invasive (pyseer)',]
names = ['emm_Family', 'emm_Sub-type', 'tissue_source', 'Clinical', 'Invasive_pyseer',]
figsizes = [(40, 2.5), (80, 2.5), (15, 2.5), (7, 2.5), (2.5, 2.5)]


for i, v in enumerate(cols):
    plt.figure(figsize=figsizes[i])
    ax = sns.boxplot(data=df1, x=v, y='Bitscore', \
                    order=df1.groupby(v)['Bitscore'].apply(np.median).sort_values(ascending=False).index.values)
    ax.set_yscale('log')
    plt.xticks(rotation=45, ha='right')
    plt.savefig('figs/' + names[i] + '.pdf', bbox_inches='tight')
    plt.show();
```
